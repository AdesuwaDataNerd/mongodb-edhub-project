```python
# Retrieve course details with instructor information
pipeline = [
    {
        '$lookup': {
            'from': 'users',
            'localField': 'instructorId',
            'foreignField': 'userId',
            'as': 'instructor'
        }
    },
    { '$unwind': '$instructor' }
]
courses = list(db.courses.aggregate(pipeline))
pd.DataFrame([{
    'title': c['title'],
    'instructor': c['instructor']['firstName'] + ' ' + c['instructor']['lastName']
} for c in courses])
```

```python
# Find students enrolled in a particular course
course_id = db.courses.find_one({})['courseId']
pipeline = [
    { '$match': {'courseId': course_id} },
    { '$lookup': {
        'from': 'users',
        'localField': 'userId',
        'foreignField': 'userId',
        'as': 'student'
    }},
    { '$unwind': '$student' }
]
enrollments = list(db.enrollments.aggregate(pipeline))
pd.DataFrame([{
    'student': e['student']['firstName'] + ' ' + e['student']['lastName'],
    'courseId': e['courseId']
} for e in enrollments])
```

```python
# Course Enrollment Statistics: Count total enrollments per course
pipeline = [
    { '$group': { '_id': '$courseId', 'total_enrollments': { '$sum': 1 } } },
    { '$lookup': { 'from': 'courses', 'localField': '_id', 'foreignField': 'courseId', 'as': 'course' } },
    { '$unwind': '$course' },
    { '$project': { 'course_title': '$course.title', 'total_enrollments': 1 } }
]
stats = list(db.enrollments.aggregate(pipeline))
pd.DataFrame(stats)[['course_title', 'total_enrollments']]
```

```python
# Student Performance Analysis: Average grade per student
pipeline = [
    { '$group': { '_id': '$userId', 'average_score': { '$avg': '$score' } } },
    { '$lookup': { 'from': 'users', 'localField': '_id', 'foreignField': 'userId', 'as': 'student' } },
    { '$unwind': '$student' },
    { '$project': { 'student_name': { '$concat': ['$student.firstName', ' ', '$student.lastName'] }, 'average_score': 1 } }
]
performance = list(db.submissions.aggregate(pipeline))
pd.DataFrame(performance)[['student_name', 'average_score']]
```

```python
# Instructor Analytics: Total students taught by each instructor
pipeline = [
    { '$lookup': { 'from': 'courses', 'localField': 'courseId', 'foreignField': 'courseId', 'as': 'course' } },
    { '$unwind': '$course' },
    { '$group': { '_id': '$course.instructorId', 'students_taught': { '$addToSet': '$userId' } } },
    { '$project': { 'instructorId': '$_id', 'total_students': { '$size': '$students_taught' } } },
    { '$lookup': { 'from': 'users', 'localField': 'instructorId', 'foreignField': 'userId', 'as': 'instructor' } },
    { '$unwind': '$instructor' },
    { '$project': { 'instructor_name': { '$concat': ['$instructor.firstName', ' ', '$instructor.lastName'] }, 'total_students': 1 } }
]
instructor_stats = list(db.enrollments.aggregate(pipeline))
pd.DataFrame(instructor_stats)[['instructor_name', 'total_students']]
```
